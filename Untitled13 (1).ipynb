```python
import os, json, numpy as np, pandas as pd, warnings
warnings.filterwarnings("ignore")

# ----------------------------
# Paths
# ----------------------------
emb_path   = "/content/drive/MyDrive/Metric_Learning_Files/metric_name_embeddings.npy"
names_path = "/content/drive/MyDrive/Metric_Learning_Files/metric_names.json"
train_path = "/content/drive/MyDrive/Metric_Learning_Files/train_data.json"
test_path  = "/content/drive/MyDrive/Metric_Learning_Files/test_data.json"
val_path   = "/content/drive/MyDrive/Metric_Learning_Files/synthetic_data_5500.json"

# ----------------------------
# Load base data
# ----------------------------
with open(names_path, "r", encoding="utf-8") as f:
    metric_name_list = json.load(f)  # list aligned with metric_emb rows

metric_emb = np.load(emb_path)
train_df = pd.read_json(train_path, lines=False)
test_df  = pd.read_json(test_path,  lines=False)
val_df   = pd.read_json(val_path,   lines=False)

print("Original Train shape:", train_df.shape)
print("Test shape          :", test_df.shape)
print("Synthetic data shape:", val_df.shape)

print("Train columns:", list(train_df.columns))
print("Test  columns:", list(test_df.columns))
print("Val   columns:", list(val_df.columns))

# ----------------------------
# Filter train & synthetic to metrics present in test (join on metric_name)
# ----------------------------
key = test_df[["metric_name"]].drop_duplicates()

train_mod = train_df.merge(key, on="metric_name", how="inner")
val_mod   = val_df.merge(key,   on="metric_name", how="inner")

val_mod = val_mod[val_mod['score'].astype(float) <= 8].reset_index(drop=True)


train_mod_path = "/content/drive/MyDrive/Metric_Learning_Files/train_mod.json"
train_mod.to_json(train_mod_path, orient="records", force_ascii=False)
print("Saved filtered train data to:", train_mod_path)

train = train_mod.reset_index(drop=True)
val   = val_mod.reset_index(drop=True)

train_df = pd.concat([train, val], axis=0)
train_df = train_df.reset_index(drop=True)
```

```python
# ----------------------------
# Downsample score = 9 points

target_n = 2000         # <<< choose how many points with score=9 you want to keep
# target_frac = 0.2     # <<< alternatively, keep a fraction instead of an absolute number

df9  = train_df[train_df["score"] == 9]
dfnz = train_df[train_df["score"] != 9]

# If using absolute count:
if len(df9) > target_n:
    df9_down = df9.sample(n=target_n, random_state=42)
else:
    df9_down = df9

# # If using fraction instead:
# df9_down = df9.sample(frac=target_frac, random_state=42)

# Combine back
train_df = pd.concat([dfnz, df9_down], axis=0).sample(frac=1, random_state=42).reset_index(drop=True)

print("Original score-9 count:", len(df9))
print("New score-9 count     :", sum(train_df["score"] == 9))
print("Final train shape     :", train_df.shape)
```

```python

# ----------------------------
# Install modeling libs
# ----------------------------
!pip install -q transformers xgboost  # xgboost not used here but handy; remove if you want
!pip install -q scikit-learn

from transformers import AutoTokenizer, AutoModel
import torch

from sklearn.kernel_ridge import KernelRidge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# ======================================
# 1) Static metric embeddings (Gemma)
# ======================================
name_to_idx = {m: i for i, m in enumerate(metric_name_list)}

def get_metric_emb_for_series(s):
    idxs = s.map(name_to_idx).values
    return metric_emb[idxs]

train_metric_static = get_metric_emb_for_series(train_df["metric_name"])
test_metric_static  = get_metric_emb_for_series(test_df["metric_name"])
val_metric_static   = get_metric_emb_for_series(val_df["metric_name"])

print("Static metric emb train:", train_metric_static.shape)
print("Static metric emb test :", test_metric_static.shape)
print("Static metric emb val  :", val_metric_static.shape)
```

```python
# ======================================
# 2) Cross-encoder model (joint metric+convo)
# ======================================
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np
from tqdm import tqdm

# Small multilingual model for speed
ce_model_name = "google/embeddinggemma-300m"

tokenizer = AutoTokenizer.from_pretrained(ce_model_name)
ce_model = AutoModel.from_pretrained(ce_model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
ce_model.to(device)
ce_model.eval()

def build_pair_text(row):
    sp = (row.get("system_prompt")  or "") if "system_prompt"  in row else ""
    up = (row.get("user_prompt")    or "") if "user_prompt"    in row else ""
    rp = (row.get("response")       or "") if "response"       in row else ""
    m  = row.get("metric_name") or ""

    text = f"Metric: {m} [SEP] System: {sp} User: {up} Assistant: {rp}"
    return text


def encode_pairs(df, batch_size=16, max_length=256):
    all_embs = []
    n = len(df)

    # tqdm progress bar
    for start in tqdm(range(0, n, batch_size), desc="Encoding CE batches"):
        batch = df.iloc[start:start+batch_size]
        texts = [build_pair_text(row) for _, row in batch.iterrows()]

        enc = tokenizer(
            texts,
            padding=True,
            truncation=True,
            max_length=max_length,
            return_tensors="pt"
        ).to(device)

        with torch.no_grad():
            outputs = ce_model(**enc)
            # If model has pooler_output, use it; else CLS token
            if hasattr(outputs, "pooler_output") and outputs.pooler_output is not None:
                embs = outputs.pooler_output
            else:
                embs = outputs.last_hidden_state[:, 0, :]  # CLS token

        all_embs.append(embs.cpu().numpy())

    return np.vstack(all_embs)


print("Encoding cross-encoder embeddings (this may take a bit)...")
cross_train = encode_pairs(train_df, batch_size=16, max_length=256)
cross_test  = encode_pairs(test_df,  batch_size=16, max_length=256)
cross_val   = encode_pairs(val_df,   batch_size=16, max_length=256)

print("Cross-encoder emb train:", cross_train.shape)
print("Cross-encoder emb test :", cross_test.shape)
print("Cross-encoder emb val  :", cross_val.shape)
```

```python

# ======================================
# 3) Build final feature matrices
#    Features = [Gemma metric emb | Cross-encoder joint emb]
# ======================================
X_train_real = np.hstack([train_metric_static, cross_train])
X_val_synth  = np.hstack([val_metric_static,   cross_val])
X_test_full  = np.hstack([test_metric_static,  cross_test])

print("X_train_real:", X_train_real.shape)
print("X_val_synth:", X_val_synth.shape)
print("X_test_full:", X_test_full.shape)

y_train_real = train_df["score"].astype(float).values
y_val_synth  = val_df["score"].astype(float).values

# Combine real + synthetic as full training pool
X_all = np.vstack([X_train_real, X_val_synth])
y_all = np.concatenate([y_train_real, y_val_synth])

print("Combined training shape:", X_all.shape, "Targets:", y_all.shape)

```

```python
# ======================================
# 4) Train/validation split
# ======================================
X_tr, X_val, y_tr, y_val = train_test_split(
    X_all,
    y_all,
    test_size=0.2,
    random_state=42,
)

print("Train split:", X_tr.shape, "Val split:", X_val.shape)

# ======================================
# 5) Kernel method: Kernel Ridge Regression (RBF)
#    Faster to tune than deep trees, expressive via kernel
# ======================================
krr_model = Pipeline([
    ("scaler", RobustScaler(with_mean=False)),
    ("krr", KernelRidge(kernel="rbf", alpha=0.69, gamma=None)),
])

krr_model.fit(X_tr, y_tr)
y_val_pred = krr_model.predict(X_val)
val_rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
print("Validation RMSE (KernelRidge RBF):", val_rmse)
```

```python

# ======================================
# 6) Train final model on all data and predict test
# ======================================
krr_model.fit(X_all, y_all)
test_preds = krr_model.predict(X_test_full)
test_preds = np.clip(test_preds, 0.0, 10.0)

# ======================================
# 7) Save submission: ID,score
# ======================================
submission = pd.DataFrame({
    "ID": np.arange(1, len(test_preds) + 1),
    "score": test_preds-1
})

```
